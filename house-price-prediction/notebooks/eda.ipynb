## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## Load Data

```python
df = pd.read_csv("housing.csv")
df.head()
```

## Basic Info

```python
df.info()
df.describe()
```

## Missing Values

```python
df.isnull().sum()
```

## Distribution Plots

```python
df.hist(figsize=(12, 8))
plt.show()
```

## Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.show()
```

## Feature vs Target

```python
df.columns
```

```python
sns.scatterplot(x="Avg. Area Income", y="Price", data=df)
plt.title("Income vs House Price")
plt.show()
```

```python
sns.scatterplot(x="Area Population", y="Price", data=df)
plt.title("Population vs Price")
plt.show()
```

## Data Cleaning

```python
df = df.dropna()
```

## Key Insights
- Income is the strongest predictor of house prices  
- Property size (number of rooms) significantly impacts pricing  
- Population density has moderate influence on pricing  
- Address feature was removed as it is non-numeric
